$A^TA$ 행렬

```python
A = np.random.randn(5,5)
ATA = A.T@A

# 고윳값 분해
L,V = np.linalg.eig(ATA)
# 특잇값 분해
U,s,VT = np.linalg.svd(ATA)


# 특잇값이 벡터 형태로 반환되므로 행렬로 변환
S = np.zeros(np.shape(ATA))
np.fill_diagonal(S, s)

# 벡터는 부호 차이가 발생해서 모두 절댓값으로 변환
print('특잇값과 고윳값 비교')
print('고윳값:', np.sort(L))
print('특잇값:', np.sort(np.diag(S)))
print('')
print('특이벡터와 고유벡터 비교')
print('고유벡터:')
print(np.sort(np.abs(V)))
print('왼쪽 특이벡터')
print(np.sort(np.abs(U)))
print('오른쪽 특이벡터')
print(np.sort(np.abs(VT.T)))
```

$A^T + A$ 행렬

```python
A = np.random.randn(5,5)
AT_A = A.T+A

# 고윳값 분해
L,V = np.linalg.eig(AT_A)
# 특잇값 분해
U,s,VT = np.linalg.svd(AT_A)


# 특잇값이 벡터 형태로 반환되므로 행렬로 변환
S = np.zeros(np.shape(AT_A))
np.fill_diagonal(S, s)

# 벡터는 부호 차이가 발생해서 모두 절댓값으로 변환
print('특잇값과 고윳값 비교')
print('고윳값:', np.sort(L))
print('특잇값:', np.sort(np.diag(S)))
print('')
print('특이벡터와 고유벡터 비교')
print('고유벡터:')
print(np.sort(np.abs(V)))
print('왼쪽 특이벡터')
print(np.sort(np.abs(U)))
print('오른쪽 특이벡터')
print(np.sort(np.abs(VT.T)))
```

*$A^T+A$의 경우 고윳값의 부호가 동일하지 않음*
> $A^TA$로 표현할 수 없는 대칭행렬이라 고윳값의 부호가 다른건가..?!

*`full_matrices=False` 파라미터를 쓰면 되겠다 !*

**Wide Matrix**

```python
# full_matrices=True
A = np.random.randn(4,10)
U,s,VT = np.linalg.svd(A, full_matrices=True)

print('size of U:', U.shape)
print('size of VT:', VT.shape)
```

```python
# full_matrices=False
A = np.random.randn(4,10)
U,s,VT = np.linalg.svd(A, full_matrices=False)

print('size of U:', U.shape)
print('size of VT:', VT.shape)
```

**Tall Matrix**

```python
# full_matrices=True
A = np.random.randn(10,4)
U,s,VT = np.linalg.svd(A, full_matrices=True)

print('size of U:', U.shape)
print('size of VT:', VT.shape)
```

```python
# full_matrices=False
A = np.random.randn(10,4)
U,s,VT = np.linalg.svd(A, full_matrices=False)

print('size of U:', U.shape)
print('size of VT:', VT.shape)
```

**Tall & Full-rank Matrix**

```python
# 왼쪽 역행렬
A = np.random.randn(5,3)
Linv = np.linalg.inv(A.T@A)@A.T

# 의사역행렬
Pinv = np.linalg.pinv(A)

print('왼쪽 역행렬:')
print(Linv)
print('의사역행렬:')
print(Pinv)
```

**Wide & Full-rank Matrix**

```python
# 오른쪽 역행렬
A = np.random.randn(3,5)
Rinv = A.T@np.linalg.inv(A@A.T)

# 의사역행렬
Pinv = np.linalg.pinv(A)

print('왼쪽 역행렬:')
print(Rinv)
print('의사역행렬:')
print(Pinv)
```
