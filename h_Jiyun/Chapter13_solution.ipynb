$A^TA$ 

```python
A = np.random.randn(5,5)
ATA = A.T@A

#  
L,V = np.linalg.eig(ATA)
#  
U,s,VT = np.linalg.svd(ATA)


#      
S = np.zeros(np.shape(ATA))
np.fill_diagonal(S, s)

print('  ')
print(':', np.sort(L))
print(':', np.sort(np.diag(S)))
print('')
print('  ')
print(':')
print(V)
print(' ')
print(U)
print(' ')
print(VT.T)
```

*`full_matrices=False`    !*

**Wide Matrix**

```python
# full_matrices=True
A = np.random.randn(4,10)
U,s,VT = np.linalg.svd(A, full_matrices=True)

print('size of U:', U.shape)
print('size of VT:', VT.shape)
```

```python
# full_matrices=False
A = np.random.randn(4,10)
U,s,VT = np.linalg.svd(A, full_matrices=False)

print('size of U:', U.shape)
print('size of VT:', VT.shape)
```

**Tall Matrix**

```python
# full_matrices=True
A = np.random.randn(10,4)
U,s,VT = np.linalg.svd(A, full_matrices=True)

print('size of U:', U.shape)
print('size of VT:', VT.shape)
```

```python
# full_matrices=False
A = np.random.randn(10,4)
U,s,VT = np.linalg.svd(A, full_matrices=False)

print('size of U:', U.shape)
print('size of VT:', VT.shape)
```

**Tall & Full-rank Matrix**

```python
#  
A = np.random.randn(5,3)
Linv = np.linalg.inv(A.T@A)@A.T

# 
Pinv = np.linalg.pinv(A)

print(' :')
print(Linv)
print(':')
print(Pinv)
```

**Wide & Full-rank Matrix**

```python
#  
A = np.random.randn(3,5)
Rinv = A.T@np.linalg.inv(A@A.T)

# 
Pinv = np.linalg.pinv(A)

print(' :')
print(Rinv)
print(':')
print(Pinv)
```
