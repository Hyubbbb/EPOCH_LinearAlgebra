### 연습 문제 5-6

(1) 더한 결과가 rank-0


```python
A = np.array([
    [1, 1],
    [-1, -1]
])

B = np.array([
    [-1, -1],
    [1, 1]
])

print(np.linalg.matrix_rank(A+B))
```

```python
np.linalg.matrix_rank(A+B) <= np.linalg.matrix_rank(A) + np.linalg.matrix_rank(B)
```

(2) 더한 결과가 rank-1

```python
A = np.array([
    [2, 4],
    [1, 2]
])

B = np.array([
    [2, 6],
    [1, 3]
])

print(np.linalg.matrix_rank(A+B))
```

```python
np.linalg.matrix_rank(A+B) <= np.linalg.matrix_rank(A) + np.linalg.matrix_rank(B)
```

(3) 더한 결과가 rank-2

```python
A = np.array([
    [2, 4],
    [1, 2]
])

B = np.array([
    [1, 3],
    [1, -2]
])

print(np.linalg.matrix_rank(A+B))
```

```python
np.linalg.matrix_rank(A+B) <= np.linalg.matrix_rank(A) + np.linalg.matrix_rank(B)
```

+) 곱셈 행렬의 계수

*rank(AB) ≤ min{rank(A), rank(B)}*

```python
# rank-0
A = np.array([
    [1, 1],
    [-1, -1]
])

B = np.array([
    [-1, -1],
    [1, 1]
])
np.linalg.matrix_rank(A@B) <= min(np.linalg.matrix_rank(A), np.linalg.matrix_rank(B))
```

```python
# rank-1
A = np.array([
    [2, 4],
    [1, 2]
])

B = np.array([
    [2, 6],
    [1, 3]
])

np.linalg.matrix_rank(A@B) <= min(np.linalg.matrix_rank(A), np.linalg.matrix_rank(B))
```

```python
# rank-2
A = np.array([
    [2, 4],
    [1, 2]
])

B = np.array([
    [1, 3],
    [1, -2]
])

np.linalg.matrix_rank(A@B) <= min(np.linalg.matrix_rank(A), np.linalg.matrix_rank(B))
```

### 연습 문제 5-7

```python
# 도저히 모르겠어서 정답 코드를 보고 이해해보았습니다......ㅠ
import matplotlib.pyplot as plt

def makeAmatrix(M,r):
  return np.random.randn(M,r) @ np.random.randn(r,M)

matSize = 20
rs = range(2,16)

Ranks = np.zeros((len(rs),len(rs),2))

for i in range(len(rs)):
  for j in range(len(rs)):

    S = makeAmatrix(matSize,rs[i]) + makeAmatrix(matSize,rs[j])
    M = makeAmatrix(matSize,rs[i]) @ makeAmatrix(matSize,rs[j])

    Ranks[i,j,0] = np.linalg.matrix_rank(S)
    Ranks[i,j,1] = np.linalg.matrix_rank(M)



## visualization
fig,axs = plt.subplots(1,2,figsize=(10,6))
s = '+@' # symbols for title

for i in range(2):

  h = axs[i].imshow(Ranks[:,:,i],vmin=np.min(rs),vmax=np.max(rs),origin='lower',
                    extent=(rs[0],rs[-1],rs[0],rs[-1]),cmap='gray')

  fig.colorbar(h,ax=axs[i],fraction=.045)
  axs[i].set_xlabel('Rank of A')
  axs[i].set_ylabel('Rank of B')
  axs[i].set_title(f'Rank of A{s[i]}B')

plt.savefig('Figure_05_09.png',dpi=300)
plt.tight_layout()
plt.show()
```
