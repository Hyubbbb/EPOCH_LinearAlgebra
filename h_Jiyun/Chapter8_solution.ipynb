## 연습 문제 8-2

### GS process

```python
V = np.random.rand(4,4)
x1 = V[:,0]
x2 = V[:,1]
x3 = V[:,2]
x4 = V[:,3]

v1 = x1
v2 = x2 - ((x2@v1)/(v1@v1))*v1
v3 = x3 - ((x3@v1)/(v1@v1))*v1 - ((x3@v2)/(v2@v2))*v2
v4 = x4 - ((x4@v1)/(v1@v1))*v1 - ((x4@v2)/(v2@v2))*v2 -((x4@v3)/(v3@v3))*v3

VQ = np.zeros((4,4))
VQ[:,0] = v1/np.linalg.norm(v1)
VQ[:,1] = v2/np.linalg.norm(v2)
VQ[:,2] = v3/np.linalg.norm(v3)
VQ[:,3] = v4/np.linalg.norm(v4)

print(VQ)

Q,R = np.linalg.qr(V)
print('')
print(Q)
```
