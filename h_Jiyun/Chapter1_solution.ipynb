### 연습문제 1-9

```python
t = np.random.rand(2)
r = np.random.rand(2)

t_parellel = (np.dot(t, r) / np.dot(r, r)) * r
t_perpendicular = t - t_parellel

print(t_parellel)
print(t_perpendicular)

round(np.dot(t_perpendicular, r), 3)
```
