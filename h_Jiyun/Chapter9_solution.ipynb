```python
import numpy as np
import scipy.linalg
```

## [Problem 9 - 3]

### matrix A and lower & upper triangular matrix L, U


```python
A = np.random.rand(5,5)
P,L,U = scipy.linalg.lu(A)

print('L: '), print(L)
print('U: '), print(U)
```

### Determinant of A

```python
np.linalg.det(A)
```

### Determination of U

```python
np.linalg.det(U)
```

>*행렬식의 크기는 일치하지만 부호가 무작위로 달라짐을 확인*

### permutation matrix(치환행렬) 적용
A = PLU

PᵀA = LU

### Determinant of PᵀA

```python
np.linalg.det(P.T@A)
```

Determinant of U

```python
np.linalg.det(U)
```

> *행렬식의 크기와 부호 모두 일치*

## [Problem 9 - 4]

### matrix A and lower & upper triangular matrix L, U

```python
A = np.random.rand(4,4)
P,L,U = scipy.linalg.lu(A)

print('L: '), print(L)
print('U: '), print(U)
```

### inverse matrix of A,U,L

```python
U_inverse = np.linalg.inv(U)
L_inverse = np.linalg.inv(L)

A_inverse = U_inverse@L_inverse@P
```

```python
A@A_inverse
```

> *A와 A_inverse가 identity matrix일 때도 있고, 아닐 때도 있음*

**SciPy 규칙**
> A = PLU

**수학 규칙**

> A = PᵀLU

```python
P_inverse = np.linalg.inv(P)
```

```python
A_inverse2 = U_inverse@L_inverse@P_inverse
```

```python
A@A_inverse2
```

> *identity matrix만 출력됨을 확인*
