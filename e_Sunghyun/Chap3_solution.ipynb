### 연습 문제 3-1

```python
import numpy as np
v1=np.array([1,2,3,4])
v2=np.array([4,3,4,1])
print(v1-v1.mean())
print(v2-v2.mean())
```

```python
def mycorrcoef(v1,v2):
    x=v1-v1.mean()
    y=v2-v2.mean()
    pearson=np.dot(x,y)/(np.linalg.norm(x)*np.linalg.norm(y))
    cossim=np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))
    return pearson, cossim
```

```python
mycorrcoef(v1,v2)
```

```python
np.corrcoef(v1,v2)
```

```python
#이미 평균 중심화가 되어있다면
v3=np.array([-1.5, -0.5,  0.5,  1.5])
v4=np.array([1.,  0.,  1., -2.])
mycorrcoef(v3,v4)#두 결과는 같다.
```

### 연습 문제 3-2

```python
import pandas as pd
pearson=[]
cosinesim=[]
offset=[]
v1=np.array([0,1,2,3])
for i in range(-50, 51):
    v2=v1+i
    pearson.append(np.corrcoef(v1,v2)[0][0])
    cosinesim.append(np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)))
    offset.append(i)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(offset, pearson)
plt.plot(offset, cosinesim)
plt.show()
```

### 연습 문제 3-5

```python
import numpy as np
import matplotlib.pyplot as plt
kernel=np.array([-1,1])
```

```python
plt.plot(kernel)
```

```python
signal=np.zeros(30)
signal[10:20]=1
```

```python
plt.plot(signal)
```

```python
edge=np.zeros(30)
for i in range(len(signal)-1):
    edge[i]=np.dot(kernel, signal[i:i+2])
```

```python
plt.plot(edge)
plt.plot(signal)
plt.show()
```
