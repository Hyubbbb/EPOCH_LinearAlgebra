```python
T=np.random.randint(-10,11, size=(4,40))
TtT=T@T.T
TtT_inv=np.linalg.inv(TtT)
print(TtT_inv@TtT)
```

```python
R=T.T@TtT_inv
print(T@R)
```

```python
# visualize! of course :)

fig,axs = plt.subplots(2,2,figsize=(10,6))

axs[0,0].imshow(T,cmap='gray')
axs[0,0].set_title('Tall matrix')

axs[0,1].imshow(R,cmap='gray')
axs[0,1].set_title('Right inverse')

axs[1,0].imshow(T@R,cmap='gray')
axs[1,0].set_title('T@R')

axs[1,1].imshow(R@T,cmap='gray')
axs[1,1].set_title('R@T')

for a in axs.flatten():
  a.set_xticks([])
  a.set_yticks([])
  
plt.tight_layout()
plt.show()
```

### 연습 문제 7-9

```python
a=int(input())
```

```python
#for문으로 돌린다.
h=np.zeros((a,a))
for i in range(a):
    for j in range(a):
        h[i,j]=1/(i+1+j+1-1)
```

```python
fig,axs = plt.subplots(1,3,figsize=(10,6))

h1=axs[0].imshow(h)
fig.colorbar(h1, ax=axs[0], fraction=0.045)
axs[0].set_title('h matrix')

h2=axs[1].imshow(np.linalg.inv(h))
fig.colorbar(h2, ax=axs[1], fraction=0.045)
axs[1].set_title('h inverse')

h3=axs[2].imshow(np.linalg.inv(h)@h)
fig.colorbar(h3, ax=axs[2], fraction=0.045)
axs[2].set_title('h inverse')
  
plt.tight_layout()
plt.show()
```

```python
E=[]
K=[]
gE=[]
gK=[]
for i in range(3,13):
    h=scipy.linalg.hilbert(i)
    hi=np.linalg.inv(h)
    edistance=h@hi-np.eye(i)
    e=np.linalg.norm(edistance)
    k=np.linalg.cond(h)
    E.append(e)
    K.append(k)
for i in range(3,13):
    g=np.random.randn(i*i).reshape(i,i)
    gi=np.linalg.inv(g)
    edistance=g@gi-np.eye(i)
    e=np.linalg.norm(edistance)
    gk=np.linalg.cond(g)
    gE.append(e)
    gK.append(gk)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(range(3,13),np.log(E),label='hil')
plt.plot(range(3,13),np.log(gE),label="random")
plt.legend()
plt.subplot(1,2,2)
plt.plot(range(3,13),np.log(K),label="hil")
plt.plot(range(3,13),np.log(gK),label="random")
plt.legend()
```
