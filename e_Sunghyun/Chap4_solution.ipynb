### 연습 문제 4-6

```python
def matmultiply(mat1,mat2):
    result=np.zeros([mat1.shape[0],mat2.shape[1]])
    for i in range(result.shape[0]):
        for j in range(result.shape[1]):
            result[i,j]=np.dot(mat1[i,:],mat2[:,j])

    return result
```

```python
mat1=np.arange(100).reshape(2,50)
mat2=np.arange(150).reshape(50,3)
print(matmultiply(mat1,mat2))
print(mat1@mat2)
```
