### [연습문제 13-5] (Shared Problem)

```python
from scipy.signal import convolve2d
```

```python
# 그림 A 재현

# create the matrix
m = 40
n = 30

# define a 2D Gaussian for smoothing
k = int((m+n)/4)
X,Y = np.meshgrid(np.linspace(-3,3,k),np.linspace(-3,3,k))
g2d = np.exp( -(X**2 + Y**2)/(k/8) )


A = convolve2d(np.random.randn(m,n),g2d,mode='same')


# SVD and create Sigma
U,s,Vt = np.linalg.svd(A)
S = np.zeros(np.shape(A))
np.fill_diagonal(S,s)


# visualize the matrices

# and show in a plot
_,axs = plt.subplots(1,4,figsize=(12,6))

axs[0].imshow(A, aspect='equal', cmap='gray', vmin=-10,vmax=10)
axs[0].set_title('A')

axs[1].imshow(U, aspect='equal', cmap='gray')
axs[1].set_title('U')

axs[2].imshow(S, aspect='equal', cmap='gray')
axs[2].set_title('S')

axs[3].imshow(Vt, aspect='equal', cmap='gray')
axs[3].set_title('Vt')

plt.tight_layout()
plt.show()
```

- 앞쪽의 특이벡터가 더 매끄럽고 뒤쪽의 특이벡터는 더 거칠다! (공간 필터링 때문)

#### 공간 필터링이란?

- 공간 필터링은 **데이터의 공간적 변화를 분석하거나 조작하는 방법**이다. 이미지를 예로 들면, 이미지의 픽셀 강도값을 조작하여 특정 성분(저주파 또는 고주파)을 강조하거나 제거하는 과정이다.
- 고주파 : 이미지에서 빠르게 변하는 영역, 즉 가장자리(경계선), 노이즈와 같은 세부 정보
- 저주파 : 천천히 변하는 부드러운 영역, 즉 배경이나 전체적인 패턴
- S(대각선 특잇값)에서 큰 특잇값은 A의 주요 구조(저주파 성분)에 기여하며, 작은 특잇값은 세부적이고 고주파적인 성분에 기여한다.

#### 평활화된 행렬 A는 무엇인가?

- 가우시안 필터라고도 하는데 가우시안 필터는 저주파 성분을 강조하고 고주파 성분을 억제하는 저주파 필터이다.
- 이미지에서 급격한 변화를 완화시켜 부드러운 효과를 준다.

```python
# 그림 B 재현

plt.figure(figsize=(12,3))
plt.plot(100*s/np.sum(s),'ks-',markersize=10)
plt.xlabel('Component index')
plt.ylabel('Variance explained (%)')
plt.title('Scree plot')
plt.show()
```

```python
# Calculate the normalized singular values (explained variance %)
explained_variance_percentages = 100 * s / np.sum(s)

# Verify if the sum of all normalized singular values is 100%
sum_of_explained_variance = np.sum(explained_variance_percentages)

# Print results
print("Normalized Singular Values (Explained Variance %):")
print(explained_variance_percentages)
print(f"Sum of Explained Variance: {sum_of_explained_variance}%")
```

- 앞에 있는 몇 개의 성분이 대부분의 분산을 설명하고 뒤쪽 성분의 분산 기여도는 매우 작다.
- 뒤쪽 성분들은 생략해도 크게 전체를 설명하는 데 문제가 없으며 차원 축소 혹은 노이즈 제거에 유용하다.

```python
# 그림 C 재현

## now show the first N "layers" separately and summed

numLayers = 4
rank1mats = np.zeros((numLayers,m,n))


# setup the figure
_,axs = plt.subplots(2,numLayers,figsize=(10,6))

# the loop
for i in range(numLayers):
    
    # create this layer
    rank1mats[i,:,:] = np.outer(U[:,i],Vt[i,:])*S[i,i]
    
    # show this layer
    axs[0,i].imshow(rank1mats[i,:,:],cmap='gray', vmin=-10,vmax=10)
    axs[0,i].set_title(f'L {i}')
    axs[0,i].set_xticks([]), axs[0,i].set_yticks([])

    # show the cumulative sum of layers
    axs[1,i].imshow(np.sum(rank1mats[:i+1,:,:],axis=0),cmap='gray', vmin=-10,vmax=10)
    axs[1,i].set_title(f'L 0:{i}')
    axs[1,i].set_xticks([]), axs[1,i].set_yticks([])


plt.tight_layout()
plt.show()
```

- L0, L1 ... L3 로 갈수록 저주파 성분에서 고주파 성분이 되어 가고 있음.
- 누적이 점점 많이 될수록 점점 A에 대한 세부 정보가 추가되고 A와 유사도가 점진적으로 증가되는 것을 확인 할 수 있음.

### [연습문제 13-7]

```python
# 높은 최대계수 행렬
A = np.random.randn(6,4)

# 명시적 왼쪽 역행렬
Linv = np.linalg.inv(A.T@A)@A.T

# 의사 역행렬
Apinv = np.linalg.pinv(A)

# 비교
print(np.round( Linv - Apinv ,5))
```

```python
# 넓은 최대계수 행렬
A = np.random.randn(4,6)

# 명시적 오른쪽 역행렬
Rinv = A.T@np.linalg.inv(A@A.T)

# 의사 역행렬
Apinv = np.linalg.pinv(A)

# compare
print(np.round( Rinv - Apinv ,5))
```
