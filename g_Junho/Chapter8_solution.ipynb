#### [연습문제 8-1]

- 난수 행렬 Q 생성하고 Q^T와 Q^-1을 계산하는 코드를 작성해서 이를 구현.
- 그런 다음 네 식이 모두 단위 행렬이 되는 것을 보여주기.(Q^TQ, QQ^T, Q^-1Q, QQ^-1)

```python
A = np.random.randn(4,4)
Q, R = np.linalg.qr(A)
QT = Q.T
Qinv = np.linalg.inv(Q)

m1 = QT @ Q
m2 = Q @ QT
m3 = Qinv @ Q
m4 = Q @ Qinv

print(m1)
print(m2)
print(m3)
print(m4)
```

#### [연습문제 8-2]

- 그람-슈미트 과정 구현 (4x4 난수 행렬 사용). 결과를 np.linalg.qr의 Q와 대조하여 확인

```python
V = np.random.randn(4,4)
Q,R = np.linalg.qr(V)

x1 = V[:,0]
x2 = V[:,1]
x3 = V[:,2]
x4 = V[:,3]

v1 = x1
v2 = x2 - (x2@v1)/(v1@v1)*v1
v3 = x3 - (x3@v1)/(v1@v1)*v1 - (x3@v2)/(v2@v2)*v2
v4 = x4 - (x4@v1)/(v1@v1)*v1 - (x4@v2)/(v2@v2)*v2 - (x4@v3)/(v3@v3)*v3

# 단위 길이로 정규화
vv1 = v1/np.linalg.norm(v1)
vv2 = v2/np.linalg.norm(v2)
vv3 = v3/np.linalg.norm(v3)
vv4 = v4/np.linalg.norm(v4)

# 행렬화
Q2 = np.column_stack((vv1,vv2,vv3,vv4))

print(Q)
print(Q2)
print(Q-Q2)

```

#### [연습문제 8-3]

- 6x6 난수 행렬의 QR 분해로부터 U라는 직교 행렬을 만든다.
- U의 QR 분해를 계산하고 R=I임을 확인한다.
- U의 각 열의 노롬을 수정한다.
- 1~6열의 노름을 10,11,...,15값의 노름으로 설정한다.
- 변조된 U 행렬을 QR 분해하여 그 R의 대각선 원소가 10에서 15인 대각 행렬인지 확인
- 이 행렬의 Q^TQ는?
- 원소 u(1,4)=0으로 설정하여 U의 직교성을 깨뜨리면 R은 어떻게 되며 그 이유는?

```python
M = np.random.randn(6,6)
U,R = np.linalg.qr(M)
UQ,UR = np.linalg.qr(U)
print(np.round(UR,3)) # UR = I

print("-----------------------------")
# U의 1열 norm 10으로 수정
for i in range(6):
  U[:,i] = (i+10)*U[:,i]

UQ,UR = np.linalg.qr(U)
# UR의 대각선 원소가 10에서 15인 대각 행렬인지 확인
print(np.round(UR,3)) # 맞음!

print("-----------------------------")

U[0][3] = 0
UQ,UR = np.linalg.qr(U)
print(np.round(UR,3))
```

#### [연습문제 8-6]

- 난수 행렬의 QR 분해를 통해 MxN 직교 행렬을 만든다.
- np.linalg.norm을 사용하여 유도된 2-노름을 계산하고 5장에서 배운 방정식을 사용하여 M의 제곱근으로 나누어 프로베니우스 노름을 계산한다.
- 두 값이 모두 1인지 확인한다.
- 무작위로 M개의 원소를 가진 열벡터 v를 생성한 다음 v와 Qv의 노름을 계산한다. 이 노름은 서로 같아야 한다.

```python
M = np.random.randint(1,10)
N = 3

A = np.random.randn(M,N)
Q,R = np.linalg.qr(A, mode='complete')

print(np.linalg.norm(Q,2))
print(np.sqrt(np.trace(Q.T @ Q)) / np.sqrt(M))

v = np.random.randn(M)


print(np.linalg.norm(v,2))
print(np.linalg.norm(Q@v,2))
```
