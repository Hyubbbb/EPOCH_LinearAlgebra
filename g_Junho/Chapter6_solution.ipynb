#### [연습 문제 6-1]

 - 공분산 행렬을 상관 행렬로 변환
 - 변환하는 과정에서 각 행렬 원소(변수 쌍 간의 공분산)를 두 변수의 분산들의 곱으로 나눈다.

 - 공식 : R = SCS (C : 공분산 행렬, S : 변수의 역 표준 편차의 대각 행렬)

```python
# 공분산 행렬 생성 (책 143p 참고)
datamean = np.mean(dataMat, axis=0)
dataMatM = dataMat - datamean
covMat = dataMatM.T @ dataMatM
covMat /= (dataMat.shape[0] - 1)
```

```python
C = covMat
S = np.diag(1/np.sqrt(np.diag(C)))

R = S @ C @ S

R
```

```python
clim = np.max(np.abs(C)) * 0.2


plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(C, cmap='gray', vmin=-clim, vmax=clim)
plt.title('Data Covariance Matrix')
plt.colorbar()
plt.subplot(1, 2, 2)
plt.imshow(R, cmap='gray', vmin=-0.5, vmax=0.5)
plt.title('Data Correlation Coefficient Matrix')
plt.colorbar()
plt.show()

```

#### [연습 문제 6-2]

- np.corrcoef() 함수를 사용해 상관 행렬을 생성한 후 행렬끼리 동일한지 확인

```python
R2 = np.corrcoef(dataMat.T)

diff = R - R2

diff
```

```python
# R, R2, diff 시각화

plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
plt.imshow(R, cmap='gray', vmin=-0.5, vmax=0.5)
plt.title('My Data Correlation Coefficient Matrix')
plt.colorbar()
plt.subplot(1, 3, 2)
plt.imshow(R2, cmap='gray', vmin=-0.5, vmax=0.5)
plt.title('Numpy Data Correlation Coefficient Matrix')
plt.colorbar()
plt.subplot(1, 3, 3)
plt.imshow(diff, cmap='gray', vmin=-0.005, vmax=0.005)
plt.title('Difference Matrix')
plt.colorbar()
plt.show()
```

#### [연습문제 6-3]

- 변환을 적용하기 전과 후의 원안의 점을 나타내기

```python
T = np.array([[1,0.5],[0,0.5]])
T
```

```python
# 20개의 점 -> 18도씩 회전
theta = np.array([18,36,54,72,90,108,126,144,162,180,198,216,234,252,270,288,306,324,342,360])

# theta별 기하학적 점 행렬 생성
original = np.array([np.cos(theta), np.sin(theta)])

transformed = T @ original

print(original.shape)
print(transformed.shape)
```

```python
# 시각화
plt.figure(figsize=(6, 6))
plt.plot(original[0], original[1], 'ko', label='Original')
plt.plot(transformed[0], transformed[1], 'bs', label='Transformed')
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.axis('equal')
plt.legend()
plt.show()
```
