#### [연습문제 7-4]
- 왼쪽 역행렬을 구할 수 있었던 원리를 활용하여 넓은 행렬에 대한 오른쪽 역행렬을 유도
- 그런 다음 넓은 행렬에 대해 [그림 7-4]를 재현
- (힌트) 왼쪽 역행렬의 코드에서 시작해 적절하게 수정

```python
M = np.random.randint(-10,11,size=(4,40))
MMt = M @ M.T
MMt_inv = linalg.inv(MMt)
R = M.T @ MMt_inv
M @ R
```

```python
# M 시각화
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(M, cmap='gray')
plt.title('Wide Matrix')

# R 시각화
plt.subplot(2,2,2)
plt.imshow(R, cmap='gray')
plt.title('Right Inverse Matrix')

# M @ R 시각화
plt.subplot(2,2,3)
plt.imshow(M @ R)
plt.title('M @ R')

# R @ M 시각화
plt.subplot(2,2,4)
plt.imshow(R @ M, cmap='gray')
plt.title('R @ M')

plt.tight_layout()
plt.show()
```

#### [연습문제 7-9]
- 힐버트 행렬 생성
- [식 7-1]에 맞춰 정수를 입력으로 받아 힐버트 행렬을 출력으로 생성하는 파이썬 함수 작성
- [그림 7-5] 재현
- (힌트) 수학 공식에 따라 행과 열(i와 j 행렬 인덱스)에 대해 이중 for 문을 사용하여 파이썬 함수를 작성하는 것이 좋음.
- (힌트) 함수가 올바른지 확인했다면 선택 사항으로 for 반복문을 사용하지 않고 함수를 다시 작성해 보세요.(힌트 : 외적)
- (힌트) 함수의 정확성은 scipy.linalg 라이브러리에 있는 힐버트 함수와 비교하여 확인할 수 있음.

```python
# 힐버트 행렬 생성
H = np.zeros((5,5))
for i in range(5):
    for j in range(5):
        H[i,j] = 1/(i+j+1)

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(H, cmap='Blues_r')
plt.colorbar()
plt.title('Hilbert Matrix')

plt.subplot(1,3,2)
H_inv = np.linalg.inv(H)
plt.imshow(H_inv, cmap='Blues_r')
plt.colorbar()
plt.title('Hilbert Matrix Inverse')

plt.subplot(1,3,3)
plt.imshow(H @ H_inv, cmap='Blues_r')
plt.colorbar()
plt.title('Hilbert Matrix @ Inverse')

plt.tight_layout()
plt.show()
```

#### [연습문제 7-10]
- 힐버트 행렬 함수를 사용하여 힐버트 행렬을 만든 다음 np.linalg.inv를 사용하여 그 역을 계산하고 두 행렬의 곱을 계산
- 이 곱은 단위 행렬과 일치해야 함.
- 즉, 이 곱과 np.eye가 생성한 실제 단위 행렬 사이의 유클리드 거리는 0이어야 함. -> 유클리드 거리 계산
- 이 코드를 3x3에서 12x12에 이르는 다양한 행렬 크기 범위에 대해 수행되도록 for 문에 넣기.
- 각 행렬 크기에 대해 유클리드 거리와 힐버트 행렬의 조건수를 저장 (조건수는 행렬에서 수칫값의 퍼짐 정도를 나타내는 척도이며, np.linalg.cond 함수 사용해 얻을 수 있음)
- 다음으로 이전 코드를 되풀이하지만 힐버트 행렬 대신 가우스 난수 행렬 사용
- 마지막으로 [그림 7-7]과 같이 모든 결과를 그래프로 표시. 그림에선 시각적 해석을 용이하게 하기 위해 거리와 조건수를 로그 척도로 그림.

```python

index = np.arange(3,13)

norms = np.zeros((2,10))

conds = np.zeros((2,10))

# 힐버트 행렬
for i in range(3,13):
    H = linalg.hilbert(i)
    Hi = linalg.inv(H)
    G = np.random.randn(i,i)
    Gi = linalg.inv(G)
    HHi = H @ Hi
    GGi = G @ Gi
    norms[0][i-3] = np.linalg.norm(HHi - np.eye(i))
    norms[1][i-3] = np.linalg.norm(GGi - np.eye(i))
    conds[0][i-3] = np.linalg.cond(H)
    conds[1][i-3] = np.linalg.cond(G)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(index, np.log(norms[0]), 'ko-', label='Hilbert')
plt.plot(index, np.log(norms[1]), 'cs-', label='Random')
plt.title('Distance from Identity Matrix')
plt.xlabel('Matrix Size')
plt.ylabel('Log Norm')
plt.legend()

plt.subplot(1,2,2)
plt.plot(index, np.log(conds[0]), 'ko-', label='Hilbert')
plt.plot(index, np.log(conds[1]), 'cs-', label='Random')
plt.title('Matrix Condition Number')
plt.xlabel('Matrix Size')
plt.ylabel('Log Condition Number')
plt.legend()

plt.tight_layout()
plt.show()
```
