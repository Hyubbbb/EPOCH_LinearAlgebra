```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg
```

### [연습 문제 10-1]

```python
# 앞 내용 예시 데이터 github 참고
numcourses = [13,4,12,3,14,13,12,9,11,7,13,11,9,2,5,7,10,0,9,7]
happiness  = [70,25,54,21,80,68,84,62,57,40,60,64,45,38,51,52,58,21,75,70]

plt.figure(figsize=(6,6))

plt.plot(numcourses,happiness,'ks',markersize=15)
plt.xlabel('Number of courses taken')
plt.ylabel('General life happiness')
plt.xlim([-1,15])
plt.ylim([0,100])
plt.grid()
plt.xticks(range(0,15,2))
plt.show()
```

```python
# 책 코드 참고 -> 예측 데이터 만들기

X = np.hstack((np.ones((20,1)),np.array(numcourses,ndmin=2).T))

# 왼쪽 역 계산
X_leftinv = np.linalg.inv(X.T@X) @ X.T

# 계수 구하기
beta = X_leftinv @ happiness
beta
```

```python
pred_happiness = X @ beta

# 잔차 구하기
diff = happiness - pred_happiness

diff
```

```python
# 잔차와 모델 예측 데이터 사이의 내적
diff @ pred_happiness
```

```python
# 잔차와 모델 예측 데이터 사이의 상관계수
np.corrcoef(diff,pred_happiness)[0,1]
```

- 상관계수가 더 작네요

### [연습문제 10-4]

```python
happiness2  = [170,25,54,21,80,68,84,62,57,40,60,64,45,38,51,52,58,21,75,70]

X = np.hstack((np.ones((20,1)),np.array(numcourses,ndmin=2).T))

# 왼쪽 역 계산
X_leftinv = np.linalg.inv(X.T@X) @ X.T

# 계수 구하기
beta = X_leftinv @ happiness2
beta
```

```python
# predicted data
pred_happiness2 = X@beta


plt.figure(figsize=(6,6))

# plot the data and predicted values
plt.plot(numcourses,happiness2,'ks',markersize=15)
plt.plot(numcourses,pred_happiness2,'o-',color=[.6,.6,.6],linewidth=3,markersize=8)

# plot the residuals (errors)
for n,y,yHat in zip(numcourses,happiness2,pred_happiness2):
  plt.plot([n,n],[y,yHat],'--',color=[.8,.8,.8],zorder=-10)

plt.xlabel('Number of courses taken')
plt.ylabel('General life happiness')
plt.xlim([-1,15])
plt.ylim([0,100])
plt.xticks(range(0,15,2))
plt.legend(['Real data','Predicted data','Residual'])
plt.title(f'SSE = {np.sum((pred_happiness2-happiness2)**2):.2f}')
plt.show()
```
