### 연습문제 1-1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 필요한 벡터 생성 및 벡터 합 연산 수행

v = np.array([1,2])
w = np.array([4,-6])
vPlusW = v+w
```

```python
# [그림 1-2] 생성

plt.arrow(0,0,v[0],v[1],head_width=.3,width=.1,color="black",length_includes_head=True)
    # 0,0 : 화살표 시작 위치
    # v[0], v[1] : 화살표 끝 위치
plt.arrow(v[0],v[1],w[0],w[1],head_width=.3,width=.1,color="orange",length_includes_head=True)
plt.arrow(0,0,vPlusW[0],vPlusW[1],head_width=.3,width=.1,color="yellow",length_includes_head=True)

plt.axis('square')
plt.axis([-6,6,-6,6])
plt.grid(linestyle='--',linewidth=.5)
plt.legend(['v','w','v+w'])
plt.show()
```

```python
# 벡터 차 연산 수행

vMinusW = v-w
```

```python
# [그림 1-2] 생성

plt.arrow(0,0,v[0],v[1],head_width=.3,width=.1,color="black",length_includes_head=True)
plt.arrow(0,0,w[0],w[1],head_width=.3,width=.1,color="orange",length_includes_head=True)
plt.arrow(w[0],w[1],vMinusW[0],vMinusW[1],head_width=.3,width=.1,color="yellow",length_includes_head=True)

plt.axis('square')
plt.axis([-6,6,-6,6])
plt.grid(linestyle='--',linewidth=.5)
plt.legend(['v','w','v-w'])
plt.show()
```

### 연습문제 1-3

```python
# 벡터를 입력으로 받아 동일한 방향의 단위벡터를 출력하는 함수

def unitVec(v):
  # norm 계산
  mu = np.linalg.norm(v)
  # 단위벡터 계산 및 반환
  return v / mu
```

```python
# 영벡터 입력
unitVec(np.zeros((3,1)))
```

### 연습문제 1-9

```python
# 2차원의 랜덤 벡터 생성
t = np.random.randn(2)
r = np.random.randn(2)

# 벡터 분해
t_para = r * (np.dot(t,r) / np.dot(r,r))  # t 벡터가 r 벡터 방향으로 평행한 성분(투영된 성분)
t_perp = t - t_para                       # t 벡터에서 r 벡터 방향으로 평행한 성분을 뺀 나머지 성분 (직교 성분)

# 이하의 두 값은 동일해야 함
print(t)
    # 실제 t 벡터
print(t_para+t_perp)
    # t 벡터의 평행 성분과 직교 성분을 합친 값
```

```python
# 직교성 확인 (내적 시 0)
print(np.dot(t_para,t_perp))
```

거의 0에 근접한 값을 가짐

```python
# 그림 1-8 재현

plt.figure(figsize=(6,6))

# t, r 벡터
plt.plot([0,t[0]],[0,t[1]],color='k',linewidth=3,label="t")
plt.plot([0,r[0]],[0,r[1]],color='orange',linewidth=3,label="r")

# 평행 성분과 직교 성분
plt.plot([0,t_para[0]],[0,t_para[1]],'k--',linewidth=3,label='t|')
plt.plot([0,t_perp[0]],[0,t_perp[1]],'k:',linewidth=3,label='t⊥')

plt.legend()
plt.show()
```
