### 연습문제 6-1

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc

%matplotlib inline

rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
np.random.seed(42)

# Generate a covariance matrix with random data
data = np.random.randn(100, 100)
covMat = np.cov(data, rowvar=False)

sd = np.sqrt(np.diag(covMat))
S = np.diag(1 / sd)

# Compute correlation matrix
corrMat = S @ covMat @ S
```

```python
clim_cov = 0.03  # color range of covariance matrix
clim_corr = 0.5  # color range of correlation matrix

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(20, 5))

# covMat
h1 = axs[0].imshow(covMat, vmin=-clim_cov, vmax=clim_cov, cmap='gray')
axs[0].set_title('데이터 공분산 행렬', fontweight='bold')

# corrMat
h2 = axs[1].imshow(corrMat, vmin=-clim_corr, vmax=clim_corr, cmap='gray')
axs[1].set_title('데이터 상관 행렬', fontweight='bold')

plt.tight_layout()
plt.show()
```

### 연습문제 6-2

```python
# NumPy's corrcoef function (transposing)
corrMat_np = np.corrcoef(data.T)
```

```python
# Plotting

fig, axs = plt.subplots(1, 3, figsize=(13, 6))

# Manual corrMat
h1 = axs[0].imshow(corrMat, vmin=-0.5, vmax=0.5, cmap='gray')
axs[0].set_title('필자의 상관 행렬', fontweight='bold')

# NumPy computed corrMat
h2 = axs[1].imshow(corrMat_np, vmin=-0.5, vmax=0.5, cmap='gray')
axs[1].set_title("NumPy의 상관 행렬", fontweight='bold')

# Difference matrix
h3 = axs[2].imshow(corrMat_np - corrMat, vmin=-0.01, vmax=0.01, cmap='gray')
axs[2].set_title('차이 행렬', fontweight='bold')

fig.colorbar(h1, ax=axs[0], fraction=.045)
fig.colorbar(h2, ax=axs[1], fraction=.045)
fig.colorbar(h3, ax=axs[2], fraction=.045)

plt.tight_layout()
plt.show()
```

### 연습문제 6-3

```python
# Define the transformation matrix T
T = np.array([[1, 0.5],
              [0, 0.5]])

# Generate points inside a circle
theta = np.linspace(0, 2 * np.pi, 100)
Opts = np.array([np.cos(theta), np.sin(theta)])

# Apply the transformation to the circle points
Tpts = T @ Opts
```

```python
# Plotting
fig, ax = plt.subplots(1, 1, figsize=(6, 6))

ax.scatter(Opts[0], Opts[1], label='원본', color='black', marker='o')
ax.scatter(Tpts[0], Tpts[1], label='변형', color='orange', marker='s')

ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.set_aspect('equal')
ax.legend()
plt.grid(True)
plt.show()
```
