## 2-1. 선형 가중 결합 구현

```python
def linear_weighted_combination(max_num, size):
    scalar_list = np.random.randint(1, max_num, size)
    vector_list = []
    # 입력받은 size 갯수만큼 벡터를 여러개 생성하고 vector_list에 넣는다
    for _ in range(size):
        vector_list.append(np.random.randint(1, max_num, size))

    print("스칼라 리스트 :", scalar_list)
    # print("벡터 리스트 :", vector)
    print("벡터 리스트 :", vector_list)

    # 선형 가중 결합 결과를 더해나갈 cur 선언 (초기값 0)
    cur = 0
    for i in range(size):
        w = scalar_list[i] * vector_list[i]
        cur += w
    
    return cur
```

```python
linear_weighted_combination(5, 3)
```

## 2-2. 벡터의 차원이 다르다면? 혹은 벡터보다 스칼라가 더 많다면?

```python
def linear_weighted_combination_more_dim(max_num, size):
    scalar_list = np.random.randint(1, max_num, size)
    vector_list = []
    # 입력받은 size 갯수만큼 벡터를 여러개 생성하고 vector_list에 넣는다
    for num in range(size+1):
        if num == size:
            vector_list.append(np.random.randint(1, max_num, size+1))
        else:
            vector_list.append(np.random.randint(1, max_num, size))

    print("스칼라 리스트 :", scalar_list)
    # print("벡터 리스트 :", vector)
    print("벡터 리스트 :", vector_list)

    # 선형 가중 결합 결과를 더해나갈 cur 선언 (초기값 0)
    cur = 0
    for i in range(size):
        w = scalar_list[i] * vector_list[i]
        cur += w
    
    return cur
```

```python
# 원래 벡터의 차원이 다르면 선형 가중 결합이 안되어야 하는데 cur 값을 뱉고있다
# 함수를 수정해야함
linear_weighted_combination_more_dim(5, 3)
```

```python
def linear_weighted_combination_more_scalar(max_num, size):
    scalar_list = np.random.randint(1, max_num, size+1)
    vector_list = []
    # 입력받은 size 갯수만큼 벡터를 여러개 생성하고 vector_list에 넣는다
    for _ in range(size):
        vector_list.append(np.random.randint(1, max_num, size))

    print("스칼라 리스트 :", scalar_list)
    # print("벡터 리스트 :", vector)
    print("벡터 리스트 :", vector_list)

    # 선형 가중 결합 결과를 더해나갈 cur 선언 (초기값 0)
    cur = 0
    for i in range(size+1):
        w = scalar_list[i] * vector_list[i]
        cur += w
    
    return cur
```

```python
linear_weighted_combination_more_scalar(5, 3)
```

## 2-3. 부분공간에 무작위로 점을 찍는다

```python
# [1, 3] 벡터
vector = np.array([1, 3])


# -4 ~ +4 사이의 균일한 분포를 갖는 무작위 숫자 100개
scalars = np.random.uniform(low=-4, high=4, size=100)
```

```python
plt.figure(figsize=(5,5))

# 기저 벡터에 무작위 스칼라를 곱하여 부분공간에 점 100개 만든다
for scalar in scalars:
    point = vector * scalar

    plt.plot(point[0], point[1], "ko")

plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid()
plt.show()
```
